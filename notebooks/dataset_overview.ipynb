```python
import pandas as pd
from langdetect import detect, LangDetectException

# Define a safe language detection function that handles exceptions and checks for meaningful content
def safe_detect(text):
    try:
        # Check if text is not just numeric or too short, you might adjust the condition based on your dataset
        if pd.notnull(text) and len(text) > 10 and any(c.isalpha() for c in text):
            return detect(text)
        else:
            return None
    except LangDetectException:
        return None

# Assuming 'no_pii_action_grievancev2_df' is your DataFrame and 'subject_content_text' is the column of interest
no_pii_action_grievancev2_df['language'] = no_pii_action_grievancev2_df['subject_content_text'].apply(safe_detect)

# Now, you can filter rows with English text or other criteria based on detected languages
english_text_df = no_pii_action_grievancev2_df[no_pii_action_grievancev2_df['language'] == 'en']
# For rows in Indian languages, you might need to check against specific language codes (like 'hi' for Hindi, 'ta' for Tamil, etc.)

# Display the first few rows to verify
english_text_df.head()
```
